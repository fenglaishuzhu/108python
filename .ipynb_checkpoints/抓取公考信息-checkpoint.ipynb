```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import requests
from bs4 import  BeautifulSoup
import pandas as pd
import  numpy as np

response=requests.get('https://ah.huatu.com/zw/liaoning/shenyangshi/2021.html')





```

```python
response.status_code
```

```python
payload=response.content
print(payload)
```

```python
soup=BeautifulSoup(payload,'html.parser')
print(soup)
```

```python

table=soup.find('table',class_='layui-table')
print(table)
```

```python

column_name=[th.text for th in table.thead.tr.find_all('th')]
column_names=pd.DataFrame(column_name).T
column_names
```

```python
data_rows=[]

for tr in table.tbody.find_all('tr'):
  row=[td.text for td in tr.find_all('td')]
  data_rows.append(row)
data_rows
```

```python
import pandas as pd1
df=pd1.DataFrame(data_rows,column_names)
```
